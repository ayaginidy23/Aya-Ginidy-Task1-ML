```python
for col in df.select_dtypes(include='object').columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
# use IQR
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    return outliers

print("Outliers using IQR method:\n")
for col in numeric_cols:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col}: {len(outliers)} outliers")
```

```python
# Here i made a loop for each column to show me the visulaztion charts for each column instead of selecting coulmns
def eda_numeric(col):
    print(f"\n📊 Column: {col} (Numeric)")

    fig, axes = plt.subplots(1, 3, figsize=(18, 4))

    # Histogram
    sns.histplot(df[col], kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f"Distribution of {col}")

    # Boxplot
    sns.boxplot(x=df[col], ax=axes[1])
    axes[1].set_title(f"Boxplot of {col}")

    # Relation with booking status
    if "booking status" in df.columns:
        sns.histplot(data=df, x=col, hue="booking status", multiple="stack", bins=30, ax=axes[2])
        axes[2].set_title(f"{col} by Booking Status")

    plt.tight_layout()
    plt.show()

def eda_categorical(col):
    print(f"\n📊 Column: {col} (Categorical)")
    print(df[col].value_counts(normalize=True).head(10))


    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Countplot
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=axes[0])
    axes[0].set_title(f"Count of {col}")
    axes[0].tick_params(axis='x', rotation=45)

    # Relation with booking status
    if "booking status" in df.columns:
        sns.countplot(data=df, x=col, hue="booking status", order=df[col].value_counts().index, ax=axes[1])
        axes[1].set_title(f"{col} vs Booking Status")
        axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

def eda_date(col):
    print(f"\n📊 Column: {col} (Date)")
    df[col] = pd.to_datetime(df[col], errors="coerce")
    print("Date range:", df[col].min(), "to", df[col].max())

    df["year"] = df[col].dt.year
    df["month"] = df[col].dt.month

    # Year and month plots side by side
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    sns.countplot(data=df, x="year", hue="booking status", ax=axes[0])
    axes[0].set_title("Bookings per Year")

    sns.countplot(data=df, x="month", hue="booking status", ax=axes[1])
    axes[1].set_title("Bookings per Month")

    plt.tight_layout()
    plt.show()

# -------------------------------
# Apply EDA
# -------------------------------
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

for col in df.columns:
    if col == "Booking_ID":
        print(f"\n🆔 Column: {col} (ID) – Check uniqueness")
        print("Unique IDs:", df[col].nunique(), "out of", len(df))
    elif col in numeric_cols:
        eda_numeric(col)
    elif col in categorical_cols:
        if "date" in col.lower():
            eda_date(col)
        else:
            eda_categorical(col)
```

```python
cols = ['P-C', 'P-not-C']
titles = {
    'P-C': " Previous Cancellations (P-C)",
    'P-not-C': " Previous Successful Stays (P-not-C)"
}

plt.figure(figsize=(14,6))

for i, col in enumerate(cols, 1):
    plt.subplot(2, len(cols), i)
    sns.histplot(df[col], bins=range(int(df[col].min()), int(df[col].max())+2),
                 color="skyblue", edgecolor="black")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")

    # --- Percentage Stacked Bar ---
    plt.subplot(2, len(cols), i+len(cols))
    prop_df = df.groupby(col)['booking status'].value_counts(normalize=True).unstack().fillna(0) * 100
    prop_df.plot(kind="bar", stacked=True, ax=plt.gca(), colormap="Pastel1", edgecolor="black")

    plt.title(f"{titles[col]} by Booking Status (%)")
    plt.xlabel(col)
    plt.ylabel("Percentage")
    plt.legend(title="Booking Status")

    # --- Add percentage labels ---
    for p in plt.gca().patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        if height > 0:  # only label non-empty bars
            plt.text(x + width/2, y + height/2, f"{height:.1f}%",
                     ha="center", va="center", fontsize=8, color="black")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(12,5))

ax = sns.countplot(x="booking status", data=df, palette="pastel", ax=axes[0])
axes[0].set_title("Count of Booking Status")
axes[0].set_xlabel("Booking Status")
axes[0].set_ylabel("Count")

# Add count labels on top of bars
for p in ax.patches:
    height = p.get_height()
    axes[0].annotate(f'{height}', (p.get_x() + p.get_width()/2, height),
                     ha='center', va='bottom', fontsize=10)

status_counts = df["booking status"].value_counts()
axes[1].pie(
    status_counts,
    labels=status_counts.index,
    autopct="%1.1f%%",
    colors=sns.color_palette("pastel"),
    startangle=90,
    wedgeprops={"edgecolor":"black"}
)
axes[1].set_title("Percentage of Booking Status")

plt.tight_layout()
plt.show()
```

```python
# Another Code add more charts
df.columns = df.columns.str.strip()

# Convert the date column to datetime objects, coercing invalid dates to NaT
df['date of reservation'] = pd.to_datetime(df['date of reservation'], errors='coerce')
# Drop rows with invalid dates
df.dropna(subset=['date of reservation'], inplace=True)


categorical_cols = ['type of meal', 'room type', 'market segment type', 'booking status', 'car parking space', 'repeated']

for col in categorical_cols:
    print(f"--- Charts for: {col} ---")
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))

    # Bar Chart
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=axes[0])
    axes[0].set_title(f'Bar Chart of {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=45)

    #  Pie Chart
    df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Pastel2', ax=axes[1])
    axes[1].set_title(f'Pie Chart of {col}')
    axes[1].set_ylabel('')

    #  Stacked Bar Chart (vs. booking status)
    if col != 'booking status':
        sns.histplot(data=df, x=col, hue='booking status', multiple='stack', shrink=0.8, ax=axes[2])
        axes[2].set_title(f'Stacked Bar Chart of {col} by Booking Status')
        axes[2].set_xlabel(col)
        axes[2].set_ylabel('Count')
        axes[2].tick_params(axis='x', rotation=45)
    else:
        fig.delaxes(axes[2])

    plt.tight_layout()
    plt.show()

# Discrete Numerical Columns:


discrete_num_cols = ['number of adults', 'number of children', 'number of weekend nights', 'number of week nights', 'P-C', 'P-not-C', 'special requests']

for col in discrete_num_cols:
    print(f"--- Charts for: {col} ---")
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))

    #  Histogram
    sns.histplot(data=df, x=col, bins=df[col].nunique(), kde=False, ax=axes[0])
    axes[0].set_title(f'Histogram of {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')

    # Bar Chart
    df[col].value_counts().sort_index().plot(kind='bar', ax=axes[1])
    axes[1].set_title(f'Bar Chart of {col} (Counts)')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Count')
    axes[1].tick_params(axis='x', rotation=0)

    # Scatter Plot vs. average price
    sns.scatterplot(data=df, x=col, y='average price', ax=axes[2])
    axes[2].set_title(f'Scatter Plot of {col} vs. Average Price')
    axes[2].set_xlabel(col)
    axes[2].set_ylabel('Average Price')

    plt.tight_layout()
    plt.show()

continuous_num_cols = ['lead time', 'average price']

for col in continuous_num_cols:
    print(f"--- Charts for: {col} ---")
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))

    #  Histogram
    sns.histplot(data=df, x=col, bins=30, kde=True, ax=axes[0])
    axes[0].set_title(f'Histogram of {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')

    #  Box Plot
    sns.boxplot(data=df, y=col, ax=axes[1])
    axes[1].set_title(f'Box Plot of {col}')
    axes[1].set_ylabel(col)

    #  KDE Plot
    sns.kdeplot(data=df, x=col, fill=True, ax=axes[2])
    axes[2].set_title(f'KDE Plot of {col}')
    axes[2].set_xlabel(col)
    axes[2].set_ylabel('Density')

    plt.tight_layout()
    plt.show()



print("--- Charts for: date of reservation ---")
fig, axes = plt.subplots(1, 3, figsize=(25, 7))

# Line Chart (Daily Bookings Over Time)
df['booking_date'] = df['date of reservation'].dt.date
daily_bookings = df.groupby('booking_date').size()
daily_bookings.plot(kind='line', ax=axes[0])
axes[0].set_title('Daily Bookings Over Time')
axes[0].set_xlabel('Date of Reservation')
axes[0].set_ylabel('Number of Bookings')
axes[0].tick_params(axis='x', rotation=45)

#  Line Chart (Monthly Average Price Over Time)
monthly_avg_price = df.set_index('date of reservation')['average price'].resample('M').mean()
monthly_avg_price.plot(kind='line', ax=axes[1])
axes[1].set_title('Monthly Average Price Over Time')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Average Price')
axes[1].tick_params(axis='x', rotation=45)

#  Heatmap (Bookings by Day of Week and Month)
df['day_of_week'] = df['date of reservation'].dt.day_name()
df['month'] = df['date of reservation'].dt.month_name()
booking_heatmap_data = df.groupby(['month', 'day_of_week']).size().unstack(fill_value=0)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
booking_heatmap_data = booking_heatmap_data.reindex(index=month_order, columns=day_order)
sns.heatmap(booking_heatmap_data, annot=True, fmt='d', cmap='YlGnBu', ax=axes[2])
axes[2].set_title('Booking Heatmap by Month and Day of Week')
axes[2].set_xlabel('Day of Week')
axes[2].set_ylabel('Month')

plt.tight_layout()
plt.show()
```
